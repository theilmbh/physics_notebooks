### Introduction

This notebook provides a simulation of the basic principles of Very Long Baseline Interferometry (VLBI). 

VLBI is radio astronomy technique that allows for high-resolution measurements of distant radio sources.  

VLBI has been in the news most recently for allowing the acquisition of the first ever image that resolved the event horizon of a black hole.

This notebook simulates a random arrangement of radio telescopes receiving signals from point sources arranged on the celestial sphere. 
These point sources are arranged to form a pattern, and the goal is to generate an image of this pattern using only the signals from the simulated telescopes.

```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm_notebook as tqdm
%matplotlib notebook
```

This function generates the signals observed at each telescope from a list of point source positions on the celestial sky.

```python
def get_signals(source_directions, receiver_positions, noise_power, frequency, times):
    c = 3e8 # Speed of Light (m/s)
    
    omega = 2*np.pi*frequency # Angular Frequency
    lam = c/frequency         # wavelength
    k = 2*np.pi / lam # Wavenumber
    
    signals = []
    for ind, recv in enumerate(receiver_positions):
        recv_signal = np.zeros(t.shape)
        for source in source_directions:
            recv_signal += np.real(np.exp(-1j * (omega*times - k*np.dot(source, recv))))
        recv_signal /= np.sqrt(np.mean(recv_signal**2))
        recv_signal += np.sqrt(noise_power)*np.random.randn(len(recv_signal))
        signals.append(recv_signal)
    return signals
```

This function computes the correlations between signals observed at each telescope and computes each baseline vector.

```python
def get_corrs_baselines(signals, poss):
    corrs = []
    bls = []
    for ind, (s1, p1) in tqdm(enumerate(zip(signals, poss))):
        for (s2, p2) in zip(signals[:ind], poss[:ind]):
            bl = p2 - p1
            c = np.mean(np.sum(np.multiply(s2, s1)))
            corrs.append(c)
            bls.append(bl)
    return corrs, bls

def plot_uv_coverage(baselines):
    fig = plt.figure()
    uvs = np.array(baselines)[:, :2]
    plt.scatter(uvs[:, 0], uvs[:, 1])
```

```python
def get_image(corrs, baselines, x, y, z, lam):
    img = np.zeros(x.shape)
    for cor, bl in tqdm(zip(corrs, baselines)):
        phase = 2*np.pi*(bl[0]*x + bl[1]*y + bl[2]*z)/lam
        img += np.real(cor*np.exp(1j*2*np.pi*(bl[0]*x + bl[1]*y + bl[2]*z)/lam))
    return img
```

```python

t = np.linspace(0, 1, 5000)
recvs = [np.array([0, 0, 0]), np.array([100, 0, 0]), np.array([0, 800, 0])]
nrec = 100
ran_range = 10000
for rec in range(nrec):
    recvs.append(np.concatenate((np.random.randint(-ran_range, ran_range, size=2), [0])))

plt.figure()
plt.scatter(np.array(recvs)[:, 0], np.array(recvs)[:, 1])
plt.show()
```

```python
src_phi = np.pi
src_theta = np.pi/4
dphi = 0.0005

ss = np.linspace(0, 2*np.pi, 120)
print(ss)
srcs = []
for s in ss[:-1]:
    x = dphi*np.cos(s) + src_phi
    y = dphi*np.sin(s) + src_theta
    srcs.append([np.cos(x)*np.sin(y), np.sin(x)*np.sin(y), np.cos(y)])

src_x = np.cos(src_phi)*np.sin(src_theta)
src_y = np.sin(src_phi)*np.sin(src_theta)
src_z = np.cos(src_theta)
src2_x = np.cos(src_phi-dphi)*np.sin(src_theta)
src2_y = np.sin(src_phi-dphi)*np.sin(src_theta)
src2_z = np.cos(src_theta)

old_srcs = [[src_x, src_y, src_z], [src2_x, src2_y, src2_z]]
#print(srcs)
frequency = 100e6

sig = get_signals(srcs, recvs, 0, frequency, t)
plt.plot(t, sig[0])
```

```python
corrs, bls = get_corrs_baselines(sig, recvs)
plot_uv_coverage(bls)
```

```python
lam = 3e8 / frequency
width_phi = 0.002 # radians
width_theta = 0.002 # radians
nphi = 100
ntheta = 100

rad2as = 3600*180 / np.pi

rphi = rad2as*np.linspace(-width_phi, width_phi, nphi)
rtheta = rad2as*np.linspace(-width_theta, width_theta, ntheta)

phi, theta= np.meshgrid(np.linspace(src_phi-width_phi, src_phi+width_phi, nphi), np.linspace(src_theta-width_theta, src_theta+width_theta, ntheta))
X = np.cos(phi)*np.sin(theta)
Y = np.sin(phi)*np.sin(theta)
Z = np.cos(theta)
img = np.divide(get_image(corrs, bls, X, Y, Z, lam), 1*np.sin(theta))
```

```python
plt.figure()
plt.imshow(img, extent = [rphi[0], rphi[-1], rtheta[0], rtheta[-1]], cmap='hot')
```
